```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
```

```python
df = pd.read_csv('college_1.csv')
df.head()
```

```python
df2 = pd.read_csv('college_2.csv')
df2.head()
```

###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
df_Total = pd.concat([df, df2], axis=0)
df_Total
```

###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


###### if  10000<codekata score<15000   (Reached_expectations.csv)



###### if  7000<codekata score<10000   (Needs_Improvement.csv)


###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
## if 10000<codekata score<15000 
df_filter1 = df_Total[(df_Total['CodeKata Score']>10000) & (df_Total['CodeKata Score']<15000)] 
# Saving CSV file 
df_filter1.to_csv('Reached_expectations.csv')
df_filter1 
```

```python
## if 7000<codekata score<10000
df_filter2 = df_Total[(df_Total['CodeKata Score']>7000) & (df_Total['CodeKata Score']<10000)] 
# Saving CSV file 
df_filter2.to_csv('Needs_Improvement.csv')
df_filter2 
```

```python
## if codekate score < 7000
df_filter3 = df_Total[df_Total['CodeKata Score']< 7000] 
# Saving CSV file 
df_filter3.to_csv('Unsatisfactory.csv')
df_filter3 
```
